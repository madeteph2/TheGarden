```python
def fetch_features(dataset,var="Rainfall"):
    attr_name = []
    # correlación de pearson
    pearson_r = []
    # valor absoluto de la correlación 
    abs_pearson_r = []
    # transformamos los arrays en un DataFrame
    
    for col in dataset.columns:
        if col != var:
            attr_name.append(col)
            abs_pearson_r.append(abs(dataset[col].corr(dataset[var])))
            pearson_r.append(dataset[col].corr(dataset[var]))
            
    features = pd.DataFrame({ 'attribute': attr_name, 'corr':pearson_r, 'abs_corr':abs_pearson_r})
    features = features.set_index('attribute')
            
    return features.sort_values(by=['abs_corr'], ascending=False)

```
